```python
#'count'

# counting the repetation of elements
lst=[10,20,200,30,20,10,90,9,10,20,10,20,200,10]

print(lst.count(10))
```

```python
#  Clear

# It delete all the elements and clearing the list
# leaves the empty list behind
# return None
# lst_name.clear()

lst=[10,20,200,30,20,10,90,9,10,20,10,20,200,10]
print(lst.clear())
print(lst)
```

```python
#'index'
#lst_name.index(ele, [start, stop])-- Returns the index of first occurance of the elements
# if the ele is not found then it return valueerror
lst=[10,20,200,30,20,10,90,9,10,20,10,20,200,10]
#print(lst.index(10))
#print(lst.index(200))
print(lst.index(200,3,5))
```

```python
#'append'

# Adding elements to the list
# list_name.append(ele)--- it adds the elements at the end of list

my_lst=[10,10.20,10+20j,'Python',None,True,[1,2,3],(1,2,3),{1:100},{100,200}]

print('Before operation')
print(my_lst,id(my_lst))
print()

# 111 in the list
print(my_lst.append(111))
print('After operation')
print(my_lst,id(my_lst))
print(my_lst)
```

```python
# 'insert',

# lst_name.insert(pos,value)
# Does not replace the value rather it insert the value at specific index and shift rest of values toword the right

my_lst=[10,10.20,10+20j,'Python',None,True,[1,2,3],(1,2,3),{1:100},{100,200}]

my_lst.insert(4,999)
print(my_lst)
```

```python
# extend',
# list_name.extend(sub_lst)-- it adds indivisual elements of sub list in the main list at the end

my_lst=[10,10.20,10+20j,'Python',None,True,[1,2,3],(1,2,3),{1:100},{100,200}]
sub_lst=[999,9,99,999]

my_lst.extend(sub_lst)
print(my_lst)
```

```python
# 'pop',

#lst_name.pop()  -- by default it removes last elements from the list
#my_lst=[10,10.20,10+20j,'Python',None,True,[1,2,3],(1,2,3),{1:100},{100,200}]
#my_lst.pop()
#print(my_lst)

#lst_name.pop(index) it removes the value at specified index
my_lst=[10,10.20,10+20j,'Python',None,True,[1,2,3],(1,2,3),{1:100},{100,200}]
print(my_lst.pop(3))
print(my_lst)
```

```python
#'remove'
# lst_name.remove() -- it removes the first occurance of the value from the list and return None
my_lst=[10,10.20,10+20j,'Python',None,True,[1,2,3],(1,2,3),{1:100},{100,200}]
print(my_lst.remove(10))
print(my_lst)
```

```python
# 'reverse',

#lst_name.reverse() -- it reverse the list
#return None
my_lst=[10,10.20,10+20j,'Python',None,True,[1,2,3],(1,2,3),{1:100},{100,200}]
print(my_lst.reverse())
print(my_lst)
```

```python
# 'sort'

# sorting -- Ascending or descending
# for sorting list uses the comparision operation
# list will be homogeneous type of element can be sorted
# return None

my_lst=[10,20,300,30,20,10,90,9,10,20,10,200,10]
print("sorting list ascending Order")
print(my_lst.sort())
print(my_lst)
print("sorting list desc Order")
print(my_lst.sort(reverse=True))
print(my_lst)
```

```python
#  'copy'

# shallow copy and deep copy -- mutable object  list, dict

# object reusability
#string  -- A -Z a-z 0 9 _
#integer -- -5 to 256
# True False None

# Alicing Giving a different name to existing object
#a=1000
#b=1000
#print(a,id(a))
#print(b,id(b))

a=1000
b=a                     # alising
print(a,id(a))
print(b,id(b))
```

```python
# shallow copy
#list_name.copy()  or list[::]
# slicing and copy method will create a new object in different memory location

a=[10,20,30,40]
b=a.copy()
print(a,id(a))
print(b,id(b))

b[0]=111
print(a,id(a))
print(b,id(b))
```

```python
a=[10,20,30,40]
b=a
print(a,id(a))
print(b,id(b))

b[0]=111
print(a,id(a))
print(b,id(b))
```

```python
# problem in shallow copy
a=[10,20,30,[40,50,60],70,80,90]
b=a[::]
#print(a,id(a))
#print(b,id(b))

#b[-1]=999
#print(a)
#print(b)

b[3][0]=444
print(a,id(a))
print(b,id(b))
```

```python
# Deepcopy
from copy import deepcopy
a=[10,20,30,[40,50,60],70,80,90]
b=deepcopy(a)
print(a,id(a))
print(b,id(b))
print('****'*29)

b[3][0]=444
print(a)
print(b)
```
