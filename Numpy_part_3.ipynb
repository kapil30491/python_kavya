```python
for row in arr2:
    for ele in row:
        print(ele)
```

```python
for i in np.nditer(arr2,order='C'):
    print(i)                        # row wise format
```

```python
for i in np.nditer(arr2,order='F'):       # colwise format
    print(i)  
```

# Concatenations of arrays


```python
# same shape and size
# alog the col wise if no of rows in all the array are same
# along rows wise if no of columns in all the arry are same
```

```python
arr1=np.random.randint(10,50,5)
arr1
```

```python
arr2=np.random.randint(10,50,5)
arr2
```

```python
np.concatenate((arr1,arr2))
```

```python
arr1=np.random.randint(10,50,15).reshape(5,3)
arr1
```

```python
arr2=np.random.randint(10,50,6).reshape(2,3)
arr2
```

```python
np.concatenate((arr1,arr2),axis=0)
```
