```python
# update 
# update is similar to extend
#main_dict.update(sub_dict)  
#it updates the main dict with elements of sub_dict
# return None

my_dict={1:100,10.20:20.30,10+20j:20+30j,'a':'Python',False:True,(1,2,3):[1,2,3]}
sub_dict={11:111,22:222,33:333}

print(my_dict.update(sub_dict))
print(my_dict)
```

```python
#'fromkeys'

#create dictinory using list of keys and same value for all keys
keys=['maths','science','english']
d=dict.fromkeys(keys,[])
print(d)
```

```python
# copy
my_dict={1:100,10.20:20.30,10+20j:20+30j,'a':'Python',False:True,(1,2,3):[1,2,3]}
dup_dict=my_dict.copy()  #shallow copy
#print(dup_dict)
dup_dict[1]=111
print(my_dict,id(my_dict))
print(dup_dict,id(dup_dict))
```

```python
my_dict={1:100,10.20:20.30,10+20j:20+30j,'a':'Python',False:True,(1,2,3):{1:[1,2,3]}}
dup_dict=my_dict.copy()
#print(dul_dict)
dup_dict[(1,2,3)][1][0]=111
print(dup_dict)
print(my_dict)
```

```python
from copy import deepcopy
my_dict={1:100,10.20:20.30,10+20j:20+30j,'a':'Python',False:True,(1,2,3):{1:[1,2,3]}}
dup_dict=deepcopy(my_dict)
#print(dul_dict)
dup_dict[(1,2,3)][1][0]=111
print(dup_dict)
print(my_dict)
```
