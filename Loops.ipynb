```python
table_num=int(input("Entet the number for which you want to write the table:"))
mul_fac=1
while mul_fac<=10:
    print(f'{table_num} X {mul_fac} = {table_num*mul_fac}')
    mul_fac=mul_fac+1
```

```python
# wap which print all the number which are divisible by 5 and 7 both  in range from 1000 to 3000

start_val=1000
while start_val<=3000:
    if (start_val%5==0) and (start_val%7==0):
        print(start_val)
    start_val+=1
```

```python
# Trasfer statement  break and continue

# if any item price is more than 500 then stop the execution and calculate the bill till there
list_price=[10,20,30,100, 900, 60, 40, 30, 500, 700, 89, 98]


total_bill=0.0
for price in list_price:
    if price>500:
        break
    else:
        total_bill+=price
print(total_bill)
```

```python
# if any item more than 500 then skip the execution and calculate the 
# total bill excluding the price which is above 500
list_price=[10,20,30,100, 900, 60, 40, 30, 500, 700, 89, 98]


total_bill=0.0
for price in list_price:
    if price>500:
        continue
    else:
        total_bill+=price
print(total_bill)
```
