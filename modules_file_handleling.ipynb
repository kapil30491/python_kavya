```python
#f.read()-- reads the content of file from current cursor to current position and returns string object
#f.read(no of char) -- read the specified no of char from current cursor position and returns string object
# f.readline() -- read the current line from current cursor position and features a string object
# f.readline(no char) -- read the specifed no of char from current cursor position in the current line
# f.readlines()--reads the whole content of the file
```

```python
# f.write('string') -- writes the 'string' to the file from current position onwords
# f.writelines(['str1','str2\n'...]) write the each string as a new line into file
```

```python
# f.tell()-- gives you current cursor position
# f.seek(pos)--used to move around file in different cursior position
```

```python
# Read mode r
# if file exists then it opens the file in read mode and cursor will be at the begining of the file
# if the file does not exist then it raises the error FileNotFoundError
# in this mode you are allowed only to read but not allowed to write into file  #UnsupportedOperation: not writable

f=open(r'C:\Users\Administrator\Desktop\Python Session\sample.txt','r')
print('current cursor position before reading',f.tell())
print(f.read())
print('current curor position after reading',f.tell())
f.close()
```

```python
# read and write mode r+
# if file exists then it opens the file in and cursor will be at the begining of the file
# if the file does not exist then it raises the error FileNotFoundError
# in this mode you are allowed only to read and write both
# in read and write mode when we write the content then it replaces the content

f=open('read_write.txt','r+')
f.write('Hello My name is swara')
f.close()
```

```python
# write mode w
# if the file exits then it will first delete the content of the file after opening file in write mode and places the cursor in the 
# beginning of  the file
# if the file does not exits then it will create a file with given name and extension
# in this mode you are allowed only to write into the file but not allowed to read the file
```

```python
f=open('read_write.txt','w')
print("After opening the file current cursor position", f.tell())
print(f.write('Hello This is file handling class'))
f.close()
```

```python
# write and read mode w+
# if the file exits then it will first delete the content of the file after opening file in write mode and places the cursor in the 
# beginning of  the file
# if the file does not exits then it will create a file with given name and extension
# in this mode you are allowed only to write and read both
```

```python
f=open('home.py','w+')
print('After opening the file, current cursor position',f.tell())
f.write('''# write and read mode w+
# if the file exits then it will first delete the content of the file after opening file in write mode and places the cursor in the 
# beginning of  the file
# if the file does not exits then it will create a file with given name and extension
# in this mode you are allowed only to write and read both ''')
print('After writing the file, current cursor position',f.tell())
f.seek(10)
print(f.read())
f.close()
```

```python
# append mode 'a'
# if file is already exists then it opens the file in append mode and cursor will be at the end of the file
# if file not exists then it will create the a file with given name and extension
# in this mode you are allowed only to append but allowed to read into file
```

```python
f=open('sample.txt','a')
print('check the cursor position',f.tell())
f.close()
```

```python
f=open('append.text','a')
print('after the opening file current cursor position is',f.tell())
f.write('\nHello')
f.seek(100)
f.write('\nHello')
f.close()
```

```python
# Append and read mode 'a+'
# if file is already exists then it opens the file in append mode and cursor will be at the end of the file
# if file not exists then it will create the a file with given name and extension
# in this mode you are allowed only to append write and read both
```
