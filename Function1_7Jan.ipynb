```python
#1 positional parameter and arguments

#postion of order matters
#the number of parameter and arguments should be match

data=['male', 'male','Female','other','Female','None','male',None, 10,'Female',1,0]


#def my_filter(inp,gen):
#    out_lst=[]
#    for ele in inp:
#        if ele in gen:
#            out_lst.append(ele)
#    return out_lst

#my_filter(data,['male','Female'])

#if we change the the order of input 
def my_filter(gen,inp):
    out_lst=[]
    for ele in inp:
        if ele in gen:
            out_lst.append(ele)
    return out_lst

my_filter(['male','Female'],data)
```

```python
#2 default parameter and arguments
#define a function which take a list of data as input and return list of 'Male' values
# default parameter should always be after the positional parameter
data=['male', 'male','Female','other','Female','None','male',None, 10,'Female',1,0]

def my_filter(inp,gen='male'):
    out_lst=[]
    for ele in inp:
        if type(ele)==str:
            if ele.lower()==gen:
                out_lst.append(ele)
    return out_lst

my_filter(data,'female')
```

```python
# keywords arguments
# position does not matters
# arguments are passed to the parameter name directly in function call
#keyword should be always come after non default argument
data=['male', 'male','Female','other','Female','None','male',None, 10,'Female',1,0]

def my_filter(inp,gen):
    out_lst=[]
    for ele in inp:
        if type(ele)==str:
            if ele.lower()==gen:
                out_lst.append(ele)
    return out_lst

my_filter(gen='female',inp=data)
```

```python
#*args
#stores the extra values in form of tuple

def form_submit(name, mob,email,*var_name):
    print(name)
    print(mob)
    print(email)
    print(var_name,type(var_name))
    print('Execution complete')
```

```python
form_submit('ABC',9876543210,'abc@gmail.com','1st cross','pune',412207)
```

```python
def my_add(a=10,*c):
    print(c)
    print(a)
    return sum(c)+a

my_add(111,20,30,40,50,60,70,80)
```
