```python
# Propertites of Dictinory
# 1 dictinory is mutalbe
st_details= {'st_1':{'st_name':'Aniket',
                     'st_age':15,
                     'st_std':'IX',
                     'st_marks':96},
             'st_2':{'st_name':'Rahul',
                     'st_age':16,
                     'st_std':'X',
                     'st_marks':96}
            }

print('Before operation')
print(st_details, id(st_details))

st_details['st_1']['st_age']=17
print()
print('after operation')
print(st_details,id(st_details))
```

```python
# Dictinory is not sequence 
#All sequence are iterable True 
#all iterable are sequence  False

#sequence elements are stored in specific order 
#you can acess this element uisng index like 0, 1, 2
#eg:

nums=[10,20,30]
print(nums[0])
```

```python
#why dictinory is not sequence
#dictinory stores key value pairs not index value
student={"name": "Raj", "age":21}
student["name"]
```

```python
# dictinory support the membership operation
st_details= {'st_1':{'st_name':'Aniket',
                     'st_age':15,
                     'st_std':'IX',
                     'st_marks':96},
             'st_2':{'st_name':'Rahul',
                     'st_age':16,
                     'st_std':'X',
                     'st_marks':96}
            }
print('Aniket' in st_details['st_1'])
print('st_name' in st_details['st_1'])
print('Aniket' in st_details['st_1']['st_name'])
```
