```python
# Shopping Bill 
# Take items & prices (list / tuple) 
# Calculate total bill + 18% GST 

items = []
prices = []
n = int(input("How many items u want to buy: "))
for i in range(n):
    item = input(f"Enter item {i+1} name: ")
    price = float(input(f"Enter price of {item}: "))
    items.append(item)
    prices.append(price)
    

total = 0
for price in prices:
    total= total + price
    gst = total * 0.18
    final_amount = total + gst

# print("\n----- SHOPPING BILL -----")
for i in range(n):
    print(items[i], "=", prices[i])

# print("-------------------------")
print("Total Amount :", total)
print("GST (18%)    :", gst)
print("Final Bill   :", final_amount)

```
