```python
# factorial(num)-->num*(num-1)*(num-2)*(num-3)....1
#5 --> 5*4*3*2*1

def my_fact(num):
    if num==1:
        return 1
    else:
        return num*(my_fact(num-1))
```

```python
my_fact(5)
```

```python
# winding
my_fact(5) -->5*(my_fact(4))
my_fact(4)-->4*(my_fact(3))
my_fact(3)-->3*(my_fact(2))
my_fact(2)-->2*(my_fact(1))
my_fact(1)-->1

# unwindig
my_fact(5)-->
24*5=120
6*4=24
2*3=6
1*2=2
```

```python
#print(1 to n)

def print_numbers(n):
    if n==0:
        return 1
    else:
        print(n)
    print_numbers(n-1)
```

```python
# Genrator 
genrator is specail function that returns once at time using the yield keyword
# Does not store all values in memory
```

```python
def my_gen():
    yield 1
    yield 2
    yield 3

g=my_gen()
print(next(g))
print(next(g))
print(next(g))
```

```python
# eevn numbner generaotr
def even_number(n):
    for i in range(2,n+1,2):
        yield i

print(list(even_number(10)))
```
