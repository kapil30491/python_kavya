```python
#Write the regular expression and use proper method which gives output as:- 'cat' in str.
import re
str1 = 'cat mat bat rat'
pattern=r'\bcat\b'

match=re.search(pattern,str1)
print(match.group())
```

```python
#Write the regular expression and use proper method which gives output as:- ['mat', 'man'] in str.
import re
str3 = 'cat mat bat rat man'
pattern=r'\bma[a-z]'
# pattern=r'\bma.\b'
# pattern=r'ma.'
match=re.findall(pattern,str3)
print(match)
```

```python
#Write the regular expression and use proper method which Retrieves all words starting with p.
#Output:- ['peter', 'per', 'picked', 'peck', 'pickled', 'peppers']
import re
str6 = "peter giper picked a peck of pickled peppers"
new_str6= re.sub(r'\bgiper\b', 'per', str6)
print('after replacement : ',new_str6)

pattern=r'\bp\w*\b'
find=re.findall(pattern,new_str6)
print(find)
```

```python
##Write the regular expression and use proper method which Retrieves name starting with 'an' or 'ak'.
#Output:- ['anil', 'akhil', 'anant', 'ankur']
import re
str16 = 'anil akhil anant abhi arun arati arundhati abhijit ankur'

# pattern=r'\b(an|ak)\w+'
# pattern=r'\b(?:an|ak)\w+'
pattern=r'(?:an|ak)\w*'

name=re.findall(pattern,str16)
print(name)
```
