```python
#'isdisjoint', 'issubset', 'issuperset', symmetric_difference',  'symmetric_difference_update'

my_set1={'abhishek@gmail.com',
        'rahul@yahoo.com',
        'krishna@yahoo.com',
        'neha@tcs.com',
        'anamika@hotmail.com',
        'aswath@rediffmail.com',
        'somya@gmail.com'}

my_set2={'George@gmail.com',
         'rahul@kavya.com',
         'jacob@yahoo.com',
         'aswath@rediffmail.com',
         'shweta@gmail.com',
         'rahul@yahoo.com'}

my_set3={'George@gmail.com',
         'rahul@kavya.com',
         'Amit@yahoo.com',
         'ravi@tcs.com',
         'jagdeesh@hotmail.com',
         'anshu@rediffmail.com',
         'shweta@gmail.com',
         'rahul@yahoo.com'}
```

```python
# union 
# my_set1.union(my_set2).union(my_set3)
#union combine elements of two or more sets without duplicates
#A U B
#my_set1.union(my_set2).union(my_set3)
my_set1.union(my_set2,my_set3)
```

```python
# intersection 
# intersction means common elements in two sets
#A âˆ© B
print(my_set1.intersection(my_set2,my_set3))
print(my_set1)
print(my_set2)
print(my_set3)
```

```python
# intersection_update
# keeping the common element and updates orignal set
#so instead of returning new set it modifies existing one

print(my_set2.intersection_update(my_set2,my_set1,my_set3))
print(my_set1)
print(my_set2)
print(my_set3)
```

```python
# 'difference', 'difference_update',

#difference 
#return the element that are
#in set A but not in set b
# it does not change orignal set-- it retunr new set

print(my_set1.difference(my_set2))
print(my_set1)
print(my_set2)
```

```python
# diffeence update
#work same like difference but
#modiefies orignal set
#remove the element that also exits in other set
print(my_set1.difference_update(my_set2))
print(my_set1)
print(my_set2)
```

```python
#'isdisjoint', 'issubset', 'issuperset', symmetric_difference',  'symmetric_difference_update'

# isdisjoint
# checks weather two sets have no common elements
#returs true if nothing has common
#returns false if one element matches

a={1,2,3}
b={4,5,6}
print(a.isdisjoint(b))
```

```python
# issubset
#checks weather all the elements of set a exist inside set b
a={1,2,4}
b={1,2,3}
print(a.issubset(b))
```

```python
# issuperset
#opposite of subset
#does set A contains everything from set b?
a={1,2,3,4}
b={2,3}
print(a.issuperset(b))
```

```python
# symmetric difference 
# return all the element which are available in both sets but discard the common elements
my_set1={'abhishek@gmail.com',
        'rahul@yahoo.com',
        'krishna@yahoo.com',
        'neha@tcs.com',
        'anamika@hotmail.com',
        'aswath@rediffmail.com',
        'somya@gmail.com'}

my_set2={'George@gmail.com',
         'rahul@kavya.com',
         'jacob@yahoo.com',
         'aswath@rediffmail.com',
         'shweta@gmail.com',
         'rahul@yahoo.com'}

my_set1.symmetric_difference(my_set2)
```
