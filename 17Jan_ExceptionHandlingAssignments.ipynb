```python
# Exception Handling – QUESTIONS ONLY (No Class)
# Q1. Number Conversion
# Write a program to:
# •	Take a number as input
# •	Convert it to integer
# •	Handle invalid input
# •	Print double of the number if valid
# •	Always print "Conversion Done"

try:
    number=input("Enter number : ")
    int(number)
    print("double",number*2)
except ValueError:
    print("invalid input")
finally:
        print("Conversion Done")
```

```python
# Q6. Simple Interest Calculator
# Write a program to:
# •	Take principal, rate, and time as input
# •	Handle invalid input
# •	Raise exception if any value is negative
# •	Print simple interest if valid
# •	Always print "Interest Calculation Done"

# Simple Interest=(P×R×T)/100
try:
    p=float(input("Enter Principal :"))
    r=float(input("Enter Rate : "))
    t=float(input("Enter Time :"))
    if p<0 or r<0 or t<0:
        raise Exception("Negative values not allowed")
   
    si= (p*r*t)/100
    print("Simple Intrest is :",si)
except ValueError:
    print("invalid input")
except Exception as e:
    print(e)
finally:
    print("Interest Calculation Done")
    
```

```python
# Q11. Calculator with Operator Check
# Write a program to:
# •	Take two numbers and an operator
# •	Handle invalid input
# •	Raise exception for invalid operator
# •	Print result if valid
# •	Always print "Calculator Finished"

try:
    num1=int(input("Enter 1st number : "))
    num2=int(input("Enter 2nd number : "))
    operation=input("Enter operation u want (+,-,*,/): ").strip()
    if operation=="+":
        result=num1+num2
    elif operation=="-":
        result=num1-num2
    elif operation=="*":
        result==num1*num2
    elif operation=="/":
        result=num1/num2
    else:
        raise ValueError(f"Invalid operator '{operation} ' Use +, -, *, /")
    
    print(f"Result: {num1} {operation} {num2} = {result}")
except ValueError as e:
    print("Invalid input:", e)

finally:
    print("Calculator Finished")
    
```

```python
# Q12. File Write Operation
# Write a program to:
# •	Take file name from user
# •	Write text into file
# •	Handle permission error
# •	Always print "File Write Completed"

try:
    filename=input("Enter file name : ")
    text = input("Enter text to write into the file: ")
    with open(filename,'w') as f:
        f.write(text)
        print(f"write completed {f}")
except PermissionError:
    print("Can not write")

except Exception as e:
    print("Permission Error",e)

finally:
        print("File Write Completed")

    
```

```python
# Q14. Temperature Conversion
# Write a program to:
# •	Take temperature in Celsius
# •	Handle invalid input
# •	Print temperature in Fahrenheit if valid
# •	Always print "Conversion Finished
 # fahrenheit = (celsius * 9/5) + 32

try:
    celsius=int(input("Enter temperature : "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"{celsius}°C is = {fahrenheit}°F")
except ValueError:
    print("Invalid Input")
finally:
    print("Conversion Finished")
```
