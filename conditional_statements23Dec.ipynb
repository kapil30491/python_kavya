```python
year=int(input("Enter the Year:"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(f'{year} is leap year')
        else:
            print(f'{year} is not leap year')
    else:
        print(f'{year} is leap Year')
else:
    print(f'{year} is not leap year')
    
```

```python
year=int(input("Enter the Year:"))
if (year%400==0) or ((year%4==0) and (year%100!=0)):
    print(f'{year} is leap year')
else:
    print(f'{year} is not leap year')

    
```

```python
# write a program that prompts user input as number of calls  and calculate
# calculate monthly telephone bills as per following rule
# minimum Rs 200 for upto 100 calls
# plus RS 0.60 per call for next 50 calls
# plus Rs 0.50 per call for next 50 calls
# plus Rs 0.40 per call for all calls beyond 200

no_of_calls=int(input("Enter the number of calls:"))
total_bills=0.00

if no_of_calls<=100:
    total_bills=200.00
    print(f'Your total bill is {total_bills} for your number of calls {no_of_calls}')
elif no_of_calls<=150:
    total_bills=200.00+(0.60*(no_of_calls-100))
    print(f'Your total bill is {total_bills} for your number of calls {no_of_calls}')
elif no_of_calls<=200:
    total_bills=200.00+(0.60*50)+(0.50*(no_of_calls-150))
    print(f'Your total bill is {total_bills} for your number of calls {no_of_calls}')
elif no_of_calls>200:
    total_bills=200.00+(0.60*50)+(0.50*50)+(0.40*(no_of_calls-200))
    print(f'Your total bill is {total_bills} for your number of calls {no_of_calls}')
    
```
